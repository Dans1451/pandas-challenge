```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
#school_data_complete.head()
```

```python
reading_score_avg=school_data_complete["reading_score"].mean()
#print(reading_score_avg)
```

```python
r_passing_count=0
readingscores=school_data_complete["reading_score"]
for m in readingscores:
    if m  >69:
        r_passing_count=r_passing_count+1
r_percent_passing=(r_passing_count/total_students)*100
#print(r_percent_passing)
```

```python
rm_passing=school_data_complete.loc[(school_data_complete["math_score"]>69) & 
                                    (school_data_complete["reading_score"]>69)]
rm_passing.head()
rm_passing_number=len(rm_passing["Student ID"])
rm_passing_percent=(rm_passing_number/total_students)*100
#print(rm_passing_percent)
```

```python
Output=pd.DataFrame([{"Total Schools":total_number_school,
       "Total Students":total_students,
       "Total Budget":total_budget,
       "Average Math Score":math_score_avg,
       "Average Reading Score":reading_score_avg,
       "% Passing Math": m_percent_passing,
       "% PAssing Reading":r_percent_passing,
       "% Overall Passing":rm_passing_percent}])
Output.head()
```

```python
school_summary=school_data[["school_name","type", "size", "budget"]]
school_summary=school_summary.set_index("school_name")

#school_summary.head()
```

```python
school_summary["Budget per student"]=(school_summary["budget"]/school_summary["size"])
#school_summary.head()
```

```python
grouped_by = school_data_complete.groupby("school_name")
school_math_total=grouped_by["math_score"].sum()
school_math_average=school_math_total/school_summary["size"]
school_summary["Average Math Score"]=school_math_average
#school_summary.head()
```

```python
school_reading_total=grouped_by["reading_score"].sum()
school_reading_average=school_reading_total/school_summary["size"]
school_summary["Average Reading Score"]=school_reading_average
#school_summary.head()
```

```python
passing_m=school_data_complete.loc[school_data_complete["math_score"]>69]

passing_math_school_sum=passing_m.groupby("school_name").count()
passing_math_school_size=school_data_complete.groupby("school_name").count()
passing_m_score_percent=(passing_math_school_sum["math_score"]/passing_math_school_size["size"])*100

school_summary["% Passing Math"]=passing_m_score_percent
#school_summary.head()
```

```python
passing_r=school_data_complete.loc[school_data_complete["reading_score"]>69]

passing_reading_school_sum=passing_r.groupby("school_name").count()
passing_reading_school_size=school_data_complete.groupby("school_name").count()
passing_r_score_percent=(passing_reading_school_sum["reading_score"]/passing_reading_school_size["size"])*100

school_summary["% Passing Reading"]=passing_r_score_percent
#school_summary.head()
```

```python
passing_o=school_data_complete.loc[(school_data_complete["reading_score"]>69) & (school_data_complete["math_score"]>69)]

passing_overall_school_sum=passing_o.groupby("school_name").count()
passing_overall_school_size=school_data_complete.groupby("school_name").count()
passing_o_score_percent=(passing_overall_school_sum["reading_score"]/passing_overall_school_size["size"])*100

school_summary["% Overall Passing"]=passing_o_score_percent
school_summary=school_summary.sort_index()
# school_summary_formatted=school_summary
# school_summary_formatted["budget"] = school_summary_formatted["budget"].map('${:,.2f}'.format)
# school_summary_formatted["Budget per student"] = school_summary_formatted["Budget per student"].map('${:,.2f}'.format)
#school_summary
```

```python
school_summary = school_summary.rename(columns={"school_name": " ", 
                                                "type":"School Type",
                                                "budget":"Total Budget",
                                    "size": "Total Students"})
#school_summary
```

```python
school_summary_formatted=school_summary.copy()
school_summary_formatted["Total Budget"] = school_summary_formatted["Total Budget"].map('${:,.2f}'.format)
school_summary_formatted["Budget per student"] = school_summary_formatted["Budget per student"].map('${:,.2f}'.format)
school_summary_formatted
```

```python
school_summary_budgeted=school_summary[["Budget per student","Average Math Score", "Average Reading Score", 
                                        "% Passing Math", "% Passing Reading", "% Overall Passing"]]
#school_summary_budgeted
```
